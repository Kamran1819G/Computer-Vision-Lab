<div style="border: 1px solid #ccc; padding: 8px; background-color: #f0f0f0; text-align: center;">
    <strong>Theory/Procedure/Algorithm</strong>
</div>

Thresholding is a basic image processing technique used for segmenting images. It converts a grayscale image into a binary image based on a threshold value.

##### 1. Simple Thresholding
Simple thresholding applies a fixed threshold value. If the pixel value is above the threshold, it is set to a maximum value (white), otherwise, it is set to zero (black).

  $$
  dst(x, y) = 
  \begin{cases} 
  \text{MaxValue}, & \text{if } src(x, y) > \text{Threshold} \\
  0, & \text{otherwise}
  \end{cases}
  $$

This method is effective when the intensity distribution of the background and foreground is uniform. However, it fails under varying lighting conditions.

##### 2. Adaptive Thresholding
Adaptive thresholding calculates the threshold for smaller regions of the image, making it effective in handling varying lighting conditions.

- **Methods of Adaptive Thresholding**:
  - **Mean Thresholding**: Uses the mean value of a local neighborhood to calculate the threshold.
  - **Gaussian Thresholding**: Uses a weighted sum of local neighborhood values (based on a Gaussian distribution) to calculate the threshold.

  $$
  dst(x, y) = 
  \begin{cases} 
  \text{MaxValue}, & \text{if } src(x, y) > \text{T}(x, y) \\
  0, & \text{otherwise}
  \end{cases}
  $$
  Where $T(x, y)$ is a locally calculated threshold value for each pixel based on its neighborhood.



```python
# Importing necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('base.jpg', 0)  # Load as grayscale

# Simple Thresholding
_, simple_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Adaptive Thresholding (Mean)
adaptive_thresh_mean = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                             cv2.THRESH_BINARY, 11, 2)

# Adaptive Thresholding (Gaussian)
adaptive_thresh_gaussian = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                                 cv2.THRESH_BINARY, 11, 2)

# Plotting the results for comparison
titles = ['Original Image', 'Simple Thresholding', 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [image, simple_thresh, adaptive_thresh_mean, adaptive_thresh_gaussian]

plt.figure(figsize=(10, 7))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()
```

##### Explanation of Code
1. **Loading the Image**: The image is loaded in grayscale mode using `cv2.imread()`.
2. **Simple Thresholding**: The function `cv2.threshold()` is used to apply simple thresholding. Pixel values above the threshold are set to `255`, and those below are set to `0`.
3. **Adaptive Thresholding**:
    - Mean Thresholding: Uses `cv2.adaptiveThreshold()` with the mean of the neighborhood to determine the local threshold.
    - Gaussian Thresholding: Similar to mean thresholding but uses a Gaussian-weighted neighborhood.
4. **Displaying Results**: The results are displayed using matplotlib for easy comparison between the original and thresholded images.

    <strong>CONCLUSION</strong>
</div>

In this experiment, we explored two thresholding techniques: simple and adaptive thresholding, using OpenCV in Python. Simple Thresholding is straightforward and works well for images with consistent lighting and a clear contrast between the foreground and background. However, it struggles in more complex environments where lighting is uneven. On the other hand, Adaptive Thresholding (both mean and Gaussian methods) dynamically calculates thresholds based on local regions of the image, making it far more robust in cases where illumination varies across the image. This method is especially useful in real-world applications where lighting conditions are not uniform. Adaptive techniques, while computationally more intensive, provide much better results in complex image processing scenarios. Thus, the choice between simple and adaptive thresholding depends on the image characteristics and the specific task at hand.
